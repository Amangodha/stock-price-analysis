```python
# For data manipulation and analysis
import pandas as pd
import numpy as np

# For visualization
import matplotlib.pyplot as plt
import seaborn as sns

# For time-series modeling
from statsmodels.tsa.arima.model import ARIMA

# For deep learning models (LSTM)
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout

# For data preprocessing
from sklearn.preprocessing import MinMaxScaler

# For accessing stock data
import yfinance as yf

# Ignore warnings (optional)
import warnings
warnings.filterwarnings('ignore')
```

```python
# Define the stock ticker symbols for FAANG
tickers = ['META', 'AAPL', 'AMZN', 'NFLX', 'GOOGL']

# Fetch 10 years of historical stock data
start_date = '2014-01-01'
end_date = '2024-01-01'

# Create an empty dictionary to store stock data
faang_data = {}

# Fetch data for each ticker
for ticker in tickers:
    data = yf.download(ticker, start=start_date, end=end_date)
    faang_data[ticker] = data
    print(f"Fetched data for {ticker}")

# Display an example of the data for one stock
faang_data['AAPL'].head()
```

```python
# Plot closing prices of all FAANG stocks
plt.figure(figsize=(12, 6))

for ticker in tickers:
    plt.plot(faang_data[ticker]['Close'], label=ticker)

plt.title("Closing Prices of FAANG Stocks (2014-2024)")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()
```

```python
# Select AAPL (Apple) stock
aapl_data = faang_data['AAPL']

# Use only the 'Close' price for modeling
aapl_close = aapl_data['Close']

# Check for missing values and drop them
aapl_close = aapl_close.dropna()

# Display the first few rows
aapl_close.head()
```

```python
# Fit the ARIMA model
model = ARIMA(aapl_close, order=(5, 1, 0))  # (p=5, d=1, q=0)
model_fit = model.fit()

# Display the model summary
print(model_fit.summary())
```

```python
# Forecast the next 30 days
forecast = model_fit.forecast(steps=30)
print("Predicted prices for the next 30 days:")
print(forecast)

# Plot the predictions
plt.figure(figsize=(10, 5))
plt.plot(aapl_close, label="Historical Prices")
plt.plot(pd.date_range(aapl_close.index[-1], periods=30, freq='B'), forecast, label="Forecast", color='red')
plt.title("Apple Stock Price Forecast (ARIMA)")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()
```

```python
# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
aapl_scaled = scaler.fit_transform(aapl_close.values.reshape(-1, 1))

# Prepare training and testing data
train_size = int(len(aapl_scaled) * 0.8)
train_data, test_data = aapl_scaled[:train_size], aapl_scaled[train_size:]

# Create sequences
def create_sequences(data, seq_length):
    x, y = [], []
    for i in range(len(data) - seq_length):
        x.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(x), np.array(y)

sequence_length = 50  # Use the past 50 days to predict the next day
x_train, y_train = create_sequences(train_data, sequence_length)
x_test, y_test = create_sequences(test_data, sequence_length)
```

```python
# Build the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(sequence_length, 1)))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(x_train, y_train, epochs=10, batch_size=32, verbose=1)
```

```python
# Predict on test data
predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)  # Convert back to original scale

# Plot the predictions
plt.figure(figsize=(12, 6))
plt.plot(aapl_close.index[-len(y_test):], scaler.inverse_transform(y_test.reshape(-1, 1)), label="Actual Prices")
plt.plot(aapl_close.index[-len(y_test):], predictions, label="Predicted Prices", color='red')
plt.title("Apple Stock Price Prediction (LSTM)")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta
import time
import threading

# Define FAANG stock tickers
tickers = ['META', 'AAPL', 'AMZN', 'NFLX', 'GOOGL']

# Define start date
start_date = '2014-01-01'  # Historical start date

# Create a global dictionary to store FAANG data
faang_data = {}

# Function to fetch and update data
def update_data():
    global faang_data
    today = datetime.today().strftime('%Y-%m-%d')  # Update today's date dynamically
    print(f"Updating data at {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

    for ticker in tickers:
        try:
            # Get data from the last fetched date or from start_date
            if ticker in faang_data and not faang_data[ticker].empty:
                last_date = faang_data[ticker].index[-1].strftime('%Y-%m-%d')
                new_data = yf.download(ticker, start=last_date, end=today)
                faang_data[ticker] = pd.concat([faang_data[ticker], new_data]).drop_duplicates()
            else:
                # Fetch full data if it's the first time
                faang_data[ticker] = yf.download(ticker, start=start_date, end=today)

            print(f"Updated data for {ticker}")
        except Exception as e:
            print(f"Failed to update data for {ticker}: {e}")

    print("Data update completed!")

# Function to schedule tasks using threading
def run_task_at_time(target_time, task_function):
    """
    Run a task function at a specific target time every day.
    :param target_time: Time in the format 'HH:MM' (24-hour clock).
    :param task_function: The function to execute at the target time.
    """
    while True:
        # Get the current time and calculate the next run time
        now = datetime.now()
        target_time_today = datetime.strptime(target_time, "%H:%M").replace(
            year=now.year, month=now.month, day=now.day
        )

        # If the target time has already passed today, schedule it for tomorrow
        if now > target_time_today:
            target_time_today += timedelta(days=1)

        # Calculate the wait time in seconds
        wait_time = (target_time_today - now).total_seconds()

        print(f"Next update scheduled at: {target_time_today.strftime('%Y-%m-%d %H:%M:%S')}")
        time.sleep(wait_time)

        # Run the task
        task_function()

# Run the task daily at 9:00 AM
target_time = "09:00"
thread = threading.Thread(target=run_task_at_time, args=(target_time, update_data))
thread.start()
```
